## Function for getting comment/tweet of a specific keyword

```
def get_comment(keyword):
  global twitter_data
  twitter_data = []
  s=sntwitter.TwitterSearchScraper(keyword)

  # Loop Through the method to get the tweet
  for i ,tweet in enumerate(s.get_items()):

    data = [
      tweet.user.username,
      tweet.rawContent,
      tweet.likeCount,
      tweet.retweetCount,
    ]
    twitter_data.append(data)

    if i > 3000:
      break
```

```
get_comment('Glo data plan')
```

## Cleaning Text

```
def clean_text(text):
    text = re.sub(r'https?:\/\/','',text)
    text = re.sub(r'-',' ',text)
    text = re.sub(r'\W+',' ',text)
    # text = re.sub(r'[A-Za-z]')
    return text
```

## Function to get Subjectivity and Polarity

```
# Create a function to get the subjectivity
#  This tell how subjective a text is

def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# Create a function to get the Polarity
# This tells how positive or negative a text is.

def getPolarity(text):
    return TextBlob(text).sentiment.polarity
```

Getting the subjectivity and polarity For Tarrif Plan

```
df_tarrif['Subjectivity'] = df_tarrif['Tweet'].apply(getSubjectivity)
df_tarrif['Polarity'] = df_tarrif['Tweet'].apply(getPolarity)
```

## Sentiment the word into category
* Creating Function
* Using Polarity Score to sentiment comment

```
# Create a function to compute negative,positive and neutral analysis

# def getAnalysis(score):
#     if score < 0:
#         return 'Negative'
#     elif score == 0:
#         return 'Ńeutral'
#     else :
#         return 'Positive'

def getAnalysis(score):

  if score < -0.7:
    return 'Very Weak'
  elif score < 0 and score >= -0.7:
    return 'Weak'
  elif score == 0:
    return 'Neutral'
  elif score > 0 and score < 0.7:
    return 'Strong'
  else:
    return 'Very Strong'
```

Getting the analysis of the comment based on the polarity score For Tarrif Plan

```
df_tarrif['Analysis'] = df_tarrif['Polarity'].apply(getAnalysis)
```
